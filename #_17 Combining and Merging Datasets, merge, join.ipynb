```python
import pandas as pd; import numpy as np
```

```python
df1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6)})
df2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(df1)
print(df2)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3=pd.DataFrame(
    {"key1":["a","b","c","c","d","e"],
     "num1":range(6)})
df4=pd.DataFrame(
    {"key2":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(df3)
print(df4)
```

```python
pd.merge(df3, df4, left_on='key1', right_on='key2')
```

```python
pd.merge(df3, df4, left_on='key1', right_on='key2', how='inner')
```

```python
print(df3)
print(df4)
```

```python
df3.merge(df4, left_on='key1', right_on='key2', how='left')
```

```python
df3.merge(df4, left_on='key1', right_on='key2', how='right')
```

```python
print(df3)
print(df4)
```

```python
df3.merge(df4, left_on='key1', right_on='key2', how='outer')
```

```python
df1=pd.DataFrame(
    {"letter":["a","a","b",
               "b","a","c"],
     "num":range(6)}) 
df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=["a","b","e"])
```

```python
# create output widgets
widget1 = widgets.Output()
widget2 = widgets.Output()
widget3 = widgets.Output()

# render in output widgets
with widget1:
    display.display(df1)
with widget2:
    display.display("   ")
with widget3:
    display.display(df2)

# create HBox
hbox = widgets.HBox([widget1, widget2, widget3])

# render hbox
hbox
```

```python
pd.merge(df1, df2, left_on='letter', right_index=True)
```

```python
pd.merge(df1, df2, left_on='letter', right_index=True, indicator=True)
```

```python
pd.merge(df1, df2, left_on='letter', right_index=True, indicator=True, how='left')
```

```python
pd.merge(df1, df2, left_on='letter', right_index=True, indicator=True, how='left', validate='m:1')
```

```python
right=pd.DataFrame(
    [[1,2],[3,4],[5,6]],
    index=["a","c","d"],
    columns=["Tom","Tim"])
left=pd.DataFrame(
    [[7,8],[9,10],[11,12],[13,14]],
    index=["a","b","e","f"],
    columns=["Sam","Kim"])
```

```python
# create output widgets
widget1 = widgets.Output()
widget2 = widgets.Output()
widget3 = widgets.Output()

# render in output widgets
with widget1:
    display.display(right)
with widget2:
    display.display("   ")
with widget3:
    display.display(left)

# create HBox
hbox = widgets.HBox([widget1, widget2, widget3])

# render hbox
hbox
```

```python
pd.merge(right, left, right_index=True, left_index=True, how='outer')
```

```python
pd.merge(right, left, right_index=True, left_index=True, how='outer', indicator=True)
```
