```python
math=pd.DataFrame({"Class":list("AB")*3,
                   "Stu":["Kim","Sam",
                          "Tim","Tom","John","Kate"],
                   "Score":[60,70,np.nan,
                            np.nan,np.nan,80]})
math
```

```python
group=math.groupby("Class")
group.mean(numeric_only=True)
```

```python
def func_(x):
    x.Score = x.Score.fillna(x.Score.mean())
    return x
```

```python
group.apply(func_)
```

```python
group["Score"].apply(lambda x: x.fillna(x.mean()))
```

```python
math
```

```python
group=math.groupby("Class")
val={"Score":100}
func1=lambda x:x.fillna(value=val)
```

```python
group.apply(func1)
```

```python
group=math.groupby("Class")
val={"A":100,"B":50}
func1=lambda f:f.fillna(value=val[f.name])
group.apply(func1)
```
